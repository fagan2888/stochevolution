```julia
include("kmr.jl");
```

```julia
p = 1/3
N = 10
epsilon = 0.01
ts_length = 10^8;
```

```julia
kmr = KMR2x2(p, N, epsilon);
```

```julia
@time simulate(kmr, ts_length, init=0);
@time simulate(kmr, ts_length, init=0);
@time simulate(kmr, ts_length, init=0);
```

```julia
x = simulate(kmr, ts_length, init=0);
```

```julia
plot_sample_path(kmr, x)
```

```julia
plot_empirical_dist(kmr, x)
```

```julia
plot_stationary_dist(kmr)
```
